```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/kumarkarna2/EDA_Notebooks/main/brain_stroke.csv')
df.head()
```

1) LINE PLOT

```python
# Plotting line chart 
plt.figure(figsize=(8,3))
sns.lineplot(data=df, x='age', y='bmi')
plt.title('Age vs BMI')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.show()
```

2) BAR PLOT

```python
# plotting BAr plot
plt.figure(figsize=(8,4))
sns.countplot(data=df, x='gender')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

3) 3D BAR PLOT

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

work_types = df['work_type'].unique()
genders = df['gender'].unique()

for i, work_type in enumerate(work_types):
    gender_data = df[df['work_type'] == work_type]['bmi']
    ax.bar(genders, gender_data, zs=i, zdir='y', alpha=0.8, label=work_type)

ax.set_xlabel('Gender')
ax.set_ylabel('Type of Work')
ax.set_zlabel('BMI Value')
plt.legend()
plt.show()
```

4) SCATTER PLOT

```python
plt.figure(figsize=(8,4))
sns.scatterplot(x='age',y='avg_glucose_level',hue='gender',data=df)
plt.title('Average Glucose Value by Age')
```

5) 3D SCATTER PLOT

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['age'], df['bmi'], df['avg_glucose_level'])
ax.set_xlabel('Age')
ax.set_ylabel('BMI')
ax.set_zlabel('Avg Glucose Level')
plt.show()
```

6) HISTOGRAM

```python
sns.histplot(data=df, x="avg_glucose_level",bins=50,hue='gender')
```

7) STACK PLOT

```python
# Generate synthetic data
np.random.seed(0)
time_intervals = np.arange(1, 11)
categories = ['Category A', 'Category B', 'Category C']
data = np.random.randint(1, 10, size=(len(categories), len(time_intervals)))

# Create stack plot
plt.figure(figsize=(10, 6))
plt.stackplot(time_intervals, data, labels=categories)
plt.legend(loc='upper left')
plt.title('Stack Plot Example')
plt.xlabel('Time Intervals')
plt.ylabel('Values')
plt.show()
```

8) PIE CHART

```python
gender_counts = df['gender'].value_counts()

plt.figure(figsize=(8, 4))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Gender Proportion')
plt.show()
```

9) SUB-PLOTS

```python
# Sub-plots: Combine multiple plots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# First subplot
sns.countplot(data=df, x='gender', ax=axs[0, 0])
axs[0, 0].set_title('Gender Distribution')

# Second subplot
sns.histplot(data=df, x='age', kde=True, ax=axs[0, 1])
axs[0, 1].set_title('Age Distribution')

# Third subplot
sns.scatterplot(data=df, x='age', y='bmi', ax=axs[1, 0])
axs[1, 0].set_title('Age vs BMI')

# Fourth subplot
sns.boxplot(data=df, x='stroke', y='bmi', ax=axs[1, 1])
axs[1, 1].set_title('BMI by Stroke')

plt.tight_layout()
plt.show()
```

11) HEATMAP

```python
# Generate a synthetic dataset
np.random.seed(0)
data = np.random.rand(10, 5)  # 10 rows, 5 columns of random numbers

# Convert to a pandas DataFrame
df = pd.DataFrame(data, columns=['A', 'B', 'C', 'D', 'E'])

# Compute correlation matrix
corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

16) WEB PLOT

```python
# Web Plot: Multivariate Data
from math import pi

# Data
categories = ['age', 'hypertension', 'heart_disease', 'avg_glucose_level', 'bmi']
N = len(categories)

# Create values and angles
values = df.iloc[0][categories].values.flatten().tolist()
values += values[:1]
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

# Plot
ax = plt.subplot(111, polar=True)
plt.xticks(angles[:-1], categories, color='grey', size=8)
ax.plot(angles, values, linewidth=2, linestyle='solid')
ax.fill(angles, values, alpha=0.4)
plt.show()
```
